Анализ по пункту 3 задания:
Предполагаем, что анализируется англоязычный текст.
Для подсчета числа гласных и согласных букв достаточно проходить текст символ за символом и проверять, обозначает ли символ гласный или согласный звук.
В случае положительного ответа увеличивать соответствующий счетчик на единицу. Поскольку число гласных мало, то можно задать только множество или строку гласных символов, а для согласных такое множество не создавать, а проверять, чтобы символ относился к алфавитным. Если символ входит в алфавит, но не относится к гласным, он будет обозначать согласную букву (хорошо, что в английском тексте нету букв типа мягкого или твердого знака, которые не обозначают ни гласные, ни согласные звуки).

Тогда план работы по пункту 3 следующий.

2. Перевести текст в нижний регистр.
3. Задать множество гласных букв.
4. Проходить текст символ за символом, проверяя его принадлежность к множеству гласных. В противном случае проверяем его принадлежность к алфавитным символам. В случае положительного ответа найденный символ будет давать согласную букву.
5. Вывести найденные значения числа согласных и гласных символов.

Загружаем текстовый файл:

```python
# import

file_name = 'GospelJohn.txt' # file with text
try:
    f = open(file_name,"r") # open file for reading
    text = f.read()         # reading file 
    f.close()               # closing file
    print('Read successfully')
except:
    print('Error reading file!')
```

Переводим текст в нижний регистр.
Задаем строку гласных букв.
Проходим по всему тексту и проверяем, гласная или согласная буква.
Выводим результат

```python
# to lower case:
text = text.lower()
vowels = 'aeiou' # vowel string
vowel_number = 0
consonant_number = 0
# find number of vowels and consonants
for c in text: 
    if c in vowels:
        vowel_number +=1
    elif c.isalpha():
        consonant_number +=1
    
print(f'Number of vowel letters: {vowel_number}')
print(f'Number of consonant letters: {consonant_number}')
```

Итак, нами был проделан анализ англоязычного текста в отношении количества гласных и согласных букв.

Выводы:
1. Целесообразно предварительно переводить текст в нижний регистр (или увеличивать набор гласных букв за счет прописных букв)
2. В англоязычном тексте размером 499 слов (2075 знаков без учета пробелов) гласных букв оказалось 738 (около 36%), согласных букв - 1234 (около 60%). Согласных букв почти в два раза больше. Знание данного соотношения для текстов на различных языках может оказаться полезным в лингвистике, прикладной криптографии и других областях, где производится анализ текстов.
